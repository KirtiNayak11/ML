```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv')
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='green',marker='+')
```

```python
x=df.iloc[:,:-1].values
```

```python
x
```

```python
y=df.iloc[:,:1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
trainin,testin,trainout,testout=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(trainin,trainout)
```

```python
ypred=model.predict(testin)
```

```python
xpred=model.predict(trainin)
```

```python
plt.scatter(trainin,trainout,color='green')
plt.plot(trainin,xpred,color='red')
plt.title("area vs price (Training dataset)")
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

```python
plt.scatter(testin,testout,color='green')
plt.plot(trainin,xpred,color='red')
plt.title("area vs price (Test Dataset) ")
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```
