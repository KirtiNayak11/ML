```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
```

```python
year=df.iloc[:,:1].values
year
```

```python
capitalIncome=df.iloc[:,1:2].values
capitalIncome
```

```python
plt.xlabel("year")
plt.ylabel("capitalIncome")
plt.title("year Vs capitalIncome")
plt.scatter(year,capitalIncome,color='green',marker='*')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
trainin,testin,trainout,testout=train_test_split(year,capitalIncome,test_size=1/7,random_state=0)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(trainin,trainout)
```

```python
testp_Income=model.predict(testin)
```

```python
trainp_Income=model.predict(trainin)
```

```python
plt.xlabel("year")
plt.ylabel("capitalIncome")
plt.title("year Vs capitalIncome (trainingDataset)")
plt.plot(trainin,trainp_Income,color='red')
plt.scatter(trainin,trainout,color='green',marker='*')
plt.show()
```

```python
plt.xlabel("year")
plt.ylabel("capitalIncome")
plt.title("year Vs capitalIncome (test Dataset)")
plt.scatter(testin,testout,color='green',marker='*')
plt.plot(trainin,trainp_Income,color='red')
plt.show()
```
