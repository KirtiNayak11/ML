```python
import torch
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/diabetes.csv")
df.head()
```

```python
x=df.iloc[:,0:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#all feature should be in range [-1,1]
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models,Sequential
from tensorflow.keras.layers import Dense,Activation,Flatten
```

```python
model=Sequential()
model.add(Dense(128,activation='relu',input_shape=(xtrain.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(xtrain, ytrain, epochs=10, batch_size=32, validation_split=0.1)
```

```python
xtest=scaler.fit_transform(xtest)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(xtest, ytest)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
import numpy as np
# Make predictions
predictions = model.predict(xtest)

# Convert predictions to binary classes (0 or 1)
binary_predictions = np.round(predictions)

# Print the predictions
print("Predictions:", binary_predictions)
```
